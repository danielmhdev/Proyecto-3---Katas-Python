1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def frecuencia_letras(cadena):
    frecuencia = {}
    cadena = cadena.lower()
    for letra in cadena:
        if letra != " ": # Ignoramos los espacios, por lo que solo entramos si no es espacio
            if letra in frecuencia:
                frecuencia[letra] += 1
            else:
                frecuencia[letra] = 1
    return frecuencia

prueba = "Estoy cansado jefe"
print (frecuencia_letras(prueba))
```

3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros.
La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
lista_palabras = ["sol", "girasol", "solar", "luna", "consola", "parasol", "soldado", "mar"] #Creo una lista de palabras con juego.

def lista_filtrada(lista, objetivo):
    resultado = []
    for palabra in lista:
        if objetivo.lower() in palabra.lower(): #lower para evitar problemas con mayus.
            resultado.append(palabra) # agrego a la nueva lista.
    return set(resultado) #set por si alguna palabra se hubiera colado más de una vez.

print(lista_filtrada(lista_palabras, "sol"))
```

4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función map().

```python
ingresos = [100, 200, 150, 300]
gastos = [10, 30, 50, 100]

def resta(ingresos,gastos):
    return ingresos - gastos
resultado = map(resta, ingresos,gastos)
print (list(resultado))
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5).
La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
lista_notas = [4,5,3,6,9,10,2,3,1,4]
def calcular_aprobado(lista_notas, nota_aprobado = 5):
    media = 0
    for nota in lista_notas:
        media += nota
    nota_media = media/len(lista_notas)
    if nota_media >= nota_aprobado:
        return "Has aprobado", nota_media #Definimos la tupla aqui.
    else:
        return "Has suspendido", nota_media

print(calcular_aprobado(lista_notas))
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
n= 5
def calcular_factorial(n):
    if n >1:
        return n * calcular_factorial(n-1)
    else:
        return 1

print(f"El factorial de {n}: es {calcular_factorial(5)}")

```

7. Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map().

```python
pasajeros = [
    ("Luis", "Gómez", 42),
    ("Marta", "Vega", 15),
    ("Pedro", "Sanz", 33)
]
def conver_string(tupla):
    return str(tupla)

resultado_lista = (map(conver_string, pasajeros))
print (list(resultado_lista))
```

8. Escribe un programa que pida al usuario dos números e intente dividirlos. Si el usuario ingresa un valor no numérico o intenta dividir por cero, maneja esas excepciones de manera adecuada y muestra un mensaje indicando si la división fue exitosa o no.


```python
def dividir_num():
    try:
        n1 = float(input("Introduce el dividendo: "))
        n2 = float(input("Introduce el divisor: "))
        resultado_div = n1/n2
    except ValueError:
        print("Numero no valido")
    except ZeroDivisionError:
        print("No se puede dividir por cero")
    else: #Si el try se ejecuta correctamente
        print(f"El resultado es: {resultado_div}")
dividir_num()
```

9. Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]. Usa la función filter().

```python
mascotas_excluidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
mascotas = ["Perro", "Mapache", "Gato", "Tigre", "Hamster", "Cocodrilo"]

def mascotas_permitidas(mascotas):
    mascotas_permitidas = list(filter(lambda x: x not in mascotas_excluidas, mascotas))
    return mascotas_permitidas

print(mascotas_permitidas(mascotas))

```

10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

```python
def calcular_promedio(lista):
    if len(lista) == 0:
        raise ValueError("Lista vacia")
    return sum(lista) / len(lista)
lista_numeros = [8,9]
try:
    print(f"El promedio es: {calcular_promedio(lista_numeros)}")
except ValueError as e:
    print(f"️Error: {e}")
except Exception as e:
    print(f"Error: {e}")
```
