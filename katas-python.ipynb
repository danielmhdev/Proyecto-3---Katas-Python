1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros.
La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
lista_palabras = ["sol", "girasol", "solar", "luna", "consola", "parasol", "soldado", "mar"] #Creo una lista de palabras con juego.

def lista_filtrada(lista, objetivo):
    resultado = []
    for palabra in lista:
        if objetivo.lower() in palabra.lower(): #lower para evitar problemas con mayus.
            resultado.append(palabra) # agrego a la nueva lista.
    return set(resultado) #set por si alguna palabra se hubiera colado más de una vez.

print(lista_filtrada(lista_palabras, "sol"))
```

5. Escribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado (por defecto 5).
La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota_aprobado. Si es así, el estado será "aprobado"; de lo contrario, "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
lista_notas = [4,5,3,6,9,10,2,3,1,4]
def calcular_aprobado(lista_notas, nota_aprobado = 5):
    media = 0
    for nota in lista_notas:
        media += nota
    nota_media = media/len(lista_notas)
    if nota_media >= nota_aprobado:
        return "Has aprobado", nota_media #Definimos la tupla aqui.
    else:
        return "Has suspendido", nota_media

print(calcular_aprobado(lista_notas))
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
n= 5
def calcular_factorial(n):
    if n >1:
        return n * calcular_factorial(n-1)
    else:
        return 1

print(f"El factorial de {n}: es {calcular_factorial(5)}")

```

29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el carácter '#' excepto los últimos cuatro.

```python
def enmascarar_text(texto):
    longitud = len(texto)
    visible = texto[-4:]
    almohadillas = "#" * (longitud-4)
    return almohadillas + visible
print(enmascarar_text("123456789"))
```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras pero en diferente orden.

```python
def son_anagramas(palabra1, palabra2):
    palabra1 = palabra1.lower()
    palabra2 = palabra2.lower()
    return sorted(palabra1) == sorted(palabra2)

print(son_anagramas("Roma", "Amor"))
```
